## 1️⃣ Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 2️⃣ Load Dataset

```python
df = pd.read_csv('../dataset/Cost_of_Living_Index_2022.csv')
df.head()
```

## 3️⃣ Dataset Information

```python
print("Dataset Shape:", df.shape)
```

```python
df.columns
```

```python
df.info()
```

## 4️⃣ Data Cleaning

### Check Missing Values

```python
df.isnull().sum()
```

### Remove Missing Values

```python
df = df.dropna()
```

### Remove Duplicate Rows

```python
df = df.drop_duplicates()
```

### Verify Cleaned Dataset

```python
print("Cleaned Dataset Shape:", df.shape)
```

## 5️⃣ Descriptive Statistics

```python
df.describe()
```

## 6️⃣ Exploratory Data Analysis (EDA)

### 📊 Chart 1: Top 10 Countries by Cost of Living

```python
top_cost = df.sort_values(by='Cost of Living Index', ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top_cost['Country'], top_cost['Cost of Living Index'])
plt.xticks(rotation=45)
plt.title('Top 10 Countries by Cost of Living Index')
plt.xlabel('Country')
plt.ylabel('Cost of Living Index')
plt.tight_layout()
plt.show()
```

### 📊 Chart 2: Top 10 Countries by Rent Index

```python
top_rent = df.sort_values(by='Rent Index', ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top_rent['Country'], top_rent['Rent Index'])
plt.xticks(rotation=45)
plt.title('Top 10 Countries by Rent Index')
plt.xlabel('Country')
plt.ylabel('Rent Index')
plt.tight_layout()
plt.show()
```

### 📊 Chart 3: Cost of Living vs Purchasing Power

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Cost of Living Index'], df['Local Purchasing Power Index'])
plt.title('Cost of Living vs Local Purchasing Power')
plt.xlabel('Cost of Living Index')
plt.ylabel('Local Purchasing Power Index')
plt.show()
```

### 📊 Chart 4: Groceries Index Distribution

```python
plt.figure(figsize=(8,5))
plt.hist(df['Groceries Index'], bins=20)
plt.title('Distribution of Groceries Index')
plt.xlabel('Groceries Index')
plt.ylabel('Frequency')
plt.show()
```

### 📊 Chart 5: Top Purchasing Power Countries

```python
top_power = df.sort_values(by='Local Purchasing Power Index', ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top_power['Country'], top_power['Local Purchasing Power Index'])
plt.xticks(rotation=45)
plt.title('Top 10 Countries by Purchasing Power')
plt.xlabel('Country')
plt.ylabel('Purchasing Power Index')
plt.tight_layout()
plt.show()
```


## 7️⃣ Key Insights

1. Bermuda and Switzerland have the highest cost of living indexes.
2. Countries with high rent generally have high living costs.
3. Purchasing power varies significantly across countries.
4. Some countries maintain strong purchasing power despite higher expenses.
5. Groceries and rent are major contributors to living costs.
